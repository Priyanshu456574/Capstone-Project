```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/gurgaon_properties_missing_value_imputation.csv')
```

```python
df.shape
```

```python
df1 = pd.read_csv('/content/gurgaon_properties_outlier_treated (1).csv')
```

```python
df.head()
```

```python
sector_luxury = df.groupby('sector')['luxury_score'].mean().sort_values(ascending=False).head(10).reset_index()
import plotly.express as px

fig = px.bar(
    sector_luxury,
    x='luxury_score',
    y='sector',
    orientation='h',
    title="Top Luxury Sectors in Gurgaon",
    color='luxury_score',
    color_continuous_scale='magma'
)

fig.update_layout(
    xaxis_title="Average Luxury Score",
    yaxis_title="Sector"
)

fig.update_yaxes(autorange="reversed")

fig.show()
```

```python
sector_price = (
    df.groupby(['sector','property_type'])['price']
    .mean()
    .reset_index()
)
fig = px.bar(
    sector_price,
    x='sector',
    y='price',
    color='property_type',
    barmode='group',
    title="Sector-wise Average Price: Flats vs Houses",
    color_discrete_map={
        'flat':'#1f77b4',
        'house':'#ff7f0e'
    }
)

fig.update_layout(
    xaxis_title="Sector",
    yaxis_title="Average Price (Crores)",
    height=600
)

fig.update_xaxes(tickangle=45)

fig.show()
```

```python
latlong = pd.read_csv('latlong.csv')
```

```python
latlong
```

```python
latlong['latitude'] = latlong['coordinates'].str.split(',').str.get(0).str.split('°').str.get(0).astype('float')
```

```python
latlong['longitude'] = latlong['coordinates'].str.split(',').str.get(1).str.split('°').str.get(0).astype('float')
```

```python
latlong.head()
```

```python
new_df = df.merge(latlong, on='sector')
```

```python
new_df.columns
```

```python
group_df = new_df.groupby('sector').mean(numeric_only=True)[['price','price_per_sqft','built_up_area','latitude','longitude']]
```

```python
group_df
```

```python

fig = px.scatter_mapbox(group_df, lat="latitude", lon="longitude", color="price_per_sqft", size='built_up_area',
                  color_continuous_scale=px.colors.cyclical.IceFire, zoom=10,
                  mapbox_style="open-street-map",text=group_df.index)
fig.show()
```

```python
wordcloud_df = df1.merge(df, left_index=True, right_index=True)[['features','sector']]
```

```python
wordcloud_df
```

```python
import ast
main = []
for item in wordcloud_df['features'].dropna().apply(ast.literal_eval):
    main.extend(item)
```

```python
main
```

```python
from wordcloud import WordCloud
```

```python
from wordcloud import WordCloud
feature_text = ' '.join(main)
```

```python
import pickle
pickle.dump(feature_text, open('feature_text.pkl','wb'))
```

```python
feature_text
```

```python
plt.rcParams["font.family"] = "Arial"

wordcloud = WordCloud(width = 800, height = 800,
                      background_color ='white',
                      stopwords = set(['s']),  # Any stopwords you'd like to exclude
                      min_font_size = 10).generate(feature_text)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
```
