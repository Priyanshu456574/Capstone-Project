```python
# Step 1: remove unwanted rows
df = df[df['price'] != 'Price on Request'].reset_index(drop=True)

# Step 2: define and apply treat_price
def treat_price(x):
    if len(x) < 2:
        return None  # handles nan
    if x[1] == 'Lac':
        return round(float(x[0]) / 100, 2)
    else:
        return round(float(x[0]), 2)

df['price'] = df['price'].astype(str).str.split(' ').apply(treat_price)
```

```python
df['price_per_sqfeet'].value_counts()
```

```python
df['price_per_sqfeet']=df['price_per_sqfeet'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype(float)
```

```python
df['floorNum'] = (
    df['floorNum']
    .str.replace('\xa0', ' ', regex=False)   # remove non-breaking spaces
    .str.strip()
    .str.replace('Lower Ground', '-1', regex=False)
    .str.replace('Basement', '-1', regex=False)
    .str.replace('Ground', '0', regex=False)
    .str.extract(r'^(-?\d+)')                # extract leading number
    [0]
    .astype(float)
    .astype('Int64')
)

print(df['floorNum'].unique())
```
